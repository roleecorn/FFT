```python
import numpy as np
from scipy.fftpack import fft,ifft
import matplotlib.pyplot as plt
import copy
import csv
```

```python
#x軸，所有資料共用
x=np.linspace(0,1,1400)
```

```python
#設定需要取樣的訊號，頻率分量只使用200
y=7*np.sin(2*np.pi*200*x) +3*np.cos(2*np.pi*200*x)
# y=7*np.sin(2*np.pi*200*x) 
y1=3*np.sin(2*np.pi*200*x)
y2=3*np.sin(2*np.pi*100*x)
```

```python
#原始資料
plt.subplot(221)
plt.plot(x[0:50],y1[0:50]) 
plt.title('Original wave with 200')
plt.subplot(222)
plt.plot(x[0:50],y2[0:50]) 
plt.title('Original wave with 100')
```

```python
yy=fft(y) 
xf = np.arange(len(yy))#快速傅立葉變換
plt.subplot(222)
plt.plot(xf,yy.real,'r')
plt.title('fft original real',color='#7A378B')
plt.subplot(221)
plt.plot(xf,yy.imag,'r')
plt.title('fft original imag',color='#7A378B')
```

```python
#逆向ifft
z=ifft(yy)
plt.subplot(221)
plt.plot(x[0:50],z[0:50]) 
plt.title('ifft of Originals fft')
```

```python
yf=abs(fft(y))    # 取絕對值
yf1=abs(fft(y))/len(x)   #歸一化處理
yf2 = yf1[range(int(len(x)/2))] #由於對稱性，只取一半區間
```

```python
xf = np.arange(len(y))  # 頻率
xf1 = copy.deepcopy(xf)
xf2 = xf[range(int(len(x)/2))] #取一半區間
xf
```

```python
#轉換結果，從原本的
plt.subplot(222)
plt.plot(xf,yf,'r')
plt.title('FFT of Mixed wave(two sides frequency range)',fontsize=7,color='#7A378B')
```

```python
plt.subplot(223)
plt.plot(xf1,yf1,'g')
plt.title('FFT of Mixed wave(normalization)',fontsize=9,color='r')
```

```python
plt.subplot(224)
plt.plot(xf2,yf2,'b')
plt.title('FFT of Mixed wave)',fontsize=10,color='#F08080')
```
